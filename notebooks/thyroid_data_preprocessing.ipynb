```python
import pandas as pd

# Step 1: Define column names
columns = ['age', 'sex', 'on thyroxine', 'query on thyroxine', 'on antithyroid medication', 'sick',
           'pregnant', 'thyroid surgery', 'I131 treatment', 'query hypothyroid', 'query hyperthyroid',
           'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych', 'TSH measured', 'TSH',
           'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U measured', 'T4U',
           'FTI measured', 'FTI', 'TBG measured', 'TBG', 'referral source', 'binaryClass']

# Step 2: Load CSV
df = pd.read_csv("../data/hypothyroid.csv", header=None, names=columns, skiprows=1)

# Step 3: Drop unneeded columns
df.drop(columns=['TSH measured', 'T3 measured', 'TT4 measured', 'T4U measured',
                 'FTI measured', 'TBG measured', 'TBG'], inplace=True)

# Step 4: Convert numeric columns to float and fill missing
numeric_cols = ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Step 5: Convert boolean-like categorical columns (t/f) to 1/0
bool_cols = ['on thyroxine', 'query on thyroxine', 'on antithyroid medication', 'sick',
             'pregnant', 'thyroid surgery', 'I131 treatment', 'query hypothyroid',
             'query hyperthyroid', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych']
df[bool_cols] = df[bool_cols].replace({'t': 1, 'f': 0}).infer_objects()

# Step 6: Encode 'sex'
df['sex'] = df['sex'].replace({'M': 1, 'F': 0})
df['sex'] = df['sex'].fillna(-1)

# Step 7: Handle target column
df['binaryClass'] = df['binaryClass'].replace({'P': 1, 'N': 0}).infer_objects()

# Step 8: Encode 'referral source'
df['referral source'] = df['referral source'].astype('category').cat.codes

# Final view
print(df.head())
```

```python
df.to_csv("../data/thyroid_cleaned.csv", index=False)
```
